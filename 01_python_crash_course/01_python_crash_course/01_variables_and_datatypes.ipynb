# Variables and Data Types

This notebook introduces basic Python variables and data types, which are the foundation for data analysis and machine learning.

```python
# Declaring variables
age = 22
height = 5.8
name = "Prasenjit"
is_student = True

print(age, type(age))
print(height, type(height))
print(name, type(name))
print(is_student, type(is_student))
```

## Explanation
- Python automatically assigns data types.
- `int` is used for integers.
- `float` represents decimal numbers.
- `str` is for text data.
- `bool` is used for True/False values.

```python
# Basic arithmetic operations
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

## Why this matters for ML
- Machine learning models work with numerical data.
- Correct data types prevent runtime errors.
- Clean variable usage improves code readability.
